# Temel İstatistiksel Analizler
Merkezi eğilim ve dağılım ölçüleri veriyi temsil ettiğinden yorum yapmayı sağlayan değerlerdir.
Betimsel istatistikler;
* Aritmetik ortalama
* Mod :
Bir seride(değişkende) en çok tekrar eden değere mod denir.
* Medyan :
Bir seride küçükten büyüğe veya büyükten küçüğe sıraladığımızda tam orta noktadan seriyi iki eşit parçaya ayıran değere medyan(ortanca)denir.
* Kartiller :
Küçükten büyüğe doğru sıralanan bir seriyi 4 parçaya ayıran değerlere kartiller denir.

 Q1 = 1/4(n+1).terim    Q3 = 3/4(n+1).terim

 Q2 = Q3-Q1(Medyan)
* Değişim Aralığı :
Max değer-Min değer(Açıklık)
* Standart sapma :
Ortalamadan olan sapmaların genel bir ölçüsüdür.
* Kovaryans :
İki değişken arasındaki ilişkinin değişkenlik ölçüsüdür.
* Korelasyon :
İki değişken arasındaki ilişkiyi,ilişkinin anlamlı olup olmadığını,ilişkinin şiddetini ve yönünü ifade eden istatiksel bir tekniktir.


-Bir değişkenin merkezi ile ilgili bilgi edinmek için merkezi eğilim ölçüleri kullanılmaktadır. Değişkenlerin merkez çevresindeki durumlarını değerlendirmek için dağılım ölçüleri kullanılır.

```python
def betimselIstatistikler(data):
  print("Değişkeninin ortalaması: "+str(data.mean()),
      "\nDeğişkenin medyanı: "+str(data.median()),
      "\nDeğişkenin modu: "+str(data.mode()[0]),
      "\nDeğişkenin standart sapması: "+str(data.std()),
      "\nDeğişkenin varyansı: "+str(data.var()),
      "\nDeğişkenin max değeri: "+str(data.max()),
      "\nDeğişkenin min değeri: "+str(data.min()),
      "\nDeğişkenin değişim aralığı: "+str(data.max()-data.min()))

  yuzdelikler = data.quantile([0.25, 0.5, 0.75])
  print("\nYüzdelikler:")
  print("1. Çeyrek:", yuzdelikler[0.25])
  print("Medyan (2. Çeyrek):", yuzdelikler[0.5])
  print("3. Çeyrek:", yuzdelikler[0.75])
```

```python
diamonds_table=diamonds["table"]
betimselIstatistikler(diamonds_table)
"""
Değişkenin standart sapma değerinin düşük çıkması verilerin ortalama etrafında dağılım gösterdiğini ifade eder.Ortalama,mod ve medyan değerlerinin birbirine yakın çıkması
ise veri setinin simetrik dağıldığını ifade eder.
"""
```

**Korelasyon Analizi**

Değişkenler arasındaki ilişki, bu ilişkinin yönü ve şiddeti ile ilgili bilgiler sağlayan istatistiksel bir yöntemdir.

-Korelasyon katsayısı, bağımlı değişken ile bağımsız değişkenler arasındaki ilişkinin gücünü gösteren bir katsayıdır. Bu katsayı, -1 ile 1 arasında değer alır ve değişkenlerin birimlerinden bağımsızdır.

-Pozitif korelasyon, değişkenlerin birlikte arttığını veya azaldığını gösterirken, negatif korelasyon ise bir değişken artarken diğerinin azaldığını ifade eder.

* +1: Mükemmel pozitif korelasyon. Bir değişkenin değeri artarken diğer değişkenin de değerinin arttığını ifade eder.
* 0: Korelasyon yok. Değişkenler arasında herhangi bir ilişki olmadığı veya çok zayıf bir ilişki olduğu anlamına gelir.
* -1: Mükemmel negatif korelasyon. Bir değişkenin değeri artarken diğer değişkenin değerinin azaldığını ifade eder.

**Aykırı Gözlem Analizi:**

Veride genel eğilimin oldukça dışına çıkan ya da diğer gözlemlerden oldukça farklı olan gözlemlere aykırı gözlem denir.
Aykırı gözlemler ; sektör bilgisi , standart sapma yaklaşımı , z skoru yaklaşımı ve boxplot(IQR) yöntemleri ile belirlenebilir.

En sık kullanılan yöntem boxplot yöntemidir. Değişken değerleri küçükten büyüğe doğru sıralanır. Çeyrekliklerine(yüzdeliklerine) yani Q1,Q3 değerlerine karşılık gelen değerler üzerinden bir eşik değer hesaplanır ve bu eşik değere göre aykırı değer tanımı yapılır.

IQR=Q3-Q1  
   
Alt eşik değer = Q1-1.5*IQR

Üst eşik değer = Q3+1.5*IQR

```python
# Aykırı Gözlem Analizi
sayisal_diamonds=diamonds.select_dtypes(include=["float64","int64"])
sayisal_diamonds.head()
```

```python
def aykiriGozlemler(dataframe): # Metot ile parametre olarak gönderilen sayısal değişkenlerdeki aykırı gözlem tespiti yapılabilir.
  Q1= dataframe.quantile(0.25)
  Q3= dataframe.quantile(0.75)
  IQR=Q3-Q1
  alt_sinir= Q1-1.5*IQR
  ust_sinir= Q3+1.5*IQR
  aykiri_df=(dataframe<alt_sinir)|(dataframe>ust_sinir)
  return dataframe[aykiri_df]
```

```python
df_depth=aykiriGozlemler(pd.DataFrame(sayisal_diamonds["depth"]))
df_depth.dropna(inplace=True) # Null olan değişkenleri siler, inplace argümanı True yapılarak yapılan değişiklik kalıcı olur.
df_depth.head()
```

Aykırı değer problemini çözmek için kullanılan yaklaşımlar;
- Silme Yaklaşımı
- Ortalama ile Doldurma
- Baskılama Yöntemi : Alt sınırdan küçük olanları alt sınıra , üst sınırdan büyük olanları üst sınıra eşitleme yöntemidir.

```python
sns.boxplot(x="carat",data=diamonds);
```

```python
# Ortalama değer ile doldurma
df_carat=sayisal_diamonds["carat"].copy()
Q1= df_carat.quantile(0.25)
Q3= df_carat.quantile(0.75)
IQR=Q3-Q1
alt_sinir= Q1-1.5*IQR
ust_sinir= Q3+1.5*IQR
df_carat[(df_carat<(alt_sinir))|(df_carat>(ust_sinir))]=df_carat.mean()
print("Alt sınır:",alt_sinir)
print("Üst sınır:",ust_sinir)
print("Table değişkeninin ortalama değeri:",df_carat.mean())
df_carat.head()
```

```python
# Baskılama yöntemi
def aykiri_gozlem_duzenleme(dataframe, col):
    df = dataframe[col].copy()
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR

    # Alt sınırdan küçük olanları alt sınıra eşitleme
    df[df < alt_sinir] = alt_sinir
    # Üst sınırdan büyük olanları üst sınıra eşitleme
    df[df > ust_sinir] = ust_sinir

    print(col,"Alt sınır:", alt_sinir)
    print(col,"Üst sınır:", ust_sinir)

    dataframe[col] = df # Veri setinde kalıcı değişiklik olması için eşitleme işlemini yaptık.
```

```python
say_col=["carat","depth","table","price","x","y","z"]
for col in say_col:
  aykiri_gozlem_duzenleme(diamonds,col)
```

**Eksik Gözlem Analizi:**

İncelenen veri setindeki gözlemlerde eksiklik olması durumunu ifade etmektedir.

Eksik gözlemlerin veri setinden direkt çıkarılabilmesi için veri setindeki eksikliğin bazı durumlarda tamamen rastlantısal olarak oluşmuş olması gerekmektedir.Eğer eksiklikler değişkenler ile ilişkili olarak çıkan yapısal problemler ile meydana gelmiş ise bu durumda yapılacak silme işlemleri veri setinde büyük kayıplara yol açabilir.Bu yüzden veri setinde eksikliğin yapısal bir eksiklik olup olmadığının bilinmesi gerekmektedir.


**Eksik Veri Türleri:**

* Tümüyle Rastlantısal Kayıp : Diğer değişkenlerden ya da yapısal bir problemden kaynaklanan tamamen rastgele oluşan gözlemlerdir.
* Rastlantısal Kayıp : Diğer değişkenlere bağlı olarak oluşabilen eksiklik türüdür.
* Rastlantısal Olmayan Kayıp : Göz ardı edilemeyecek olan ve yapısal problemler ile ortaya çıkan eksiklik türüdür.


-Eksik veri problemi silme , değer atama ve tahmine dayalı yöntemler ile çözülebilir.

```python
# Eksik Veri Analizi
diamonds.isnull().sum()  # Veride her değişkendeki eksik gözlem sayısını verir.
```

```python
diamonds[diamonds.isnull().any(axis=1)]  # Eksik veri bulunan satırları getirir.(Bu veri setinde eksik değer yoktur.)
```

```python
diamonds.dropna(axis=1,how="all")  # Eksik veri bulunan sütundaki verilerin hepsi NAN ise silme işlemini gerçekleştirir.(axis=0 olursa satır bazında silme işlemi gerçekleştirir.)
diamonds.dropna(axis=1,how="all",inplace=True) # inlace=True yapılarak yapılan değişiklikler kalıcı olur.
```

```python
# Basit değer atama yöntemleri: Değişkendeki eksik verileri değişkenin ortalaması ile doldurma yöntemidir.
diamonds["carat"].fillna(diamonds["carat"].mean())  # carat değişkenindeki eksik verileri ortalaması ile doldurur.(Eksik veri olmadığından bu işlem gerçekleşmez.)
```

**Değişken Dönüşümleri:**

Değişken dönüşümü, veri analizi ve istatistikte, verilerin daha uygun bir formata dönüştürülmesi işlemidir. Bu dönüşüm, veriler üzerinde daha etkili analiz yapabilmek, veri setindeki ilişkileri daha iyi gözlemleyebilmek ve bazı istatistiksel yöntemlerin varsayımlarını karşılamak için yapılır.

* Label Encoding:

  Kategorik değişkenleri sayısal değerlere dönüştürmek için kullanılan bir veri dönüşüm yöntemidir. Bu yöntem, özellikle makine öğrenimi modellerinde kategorik değişkenleri sayısal olarak ifade etmek amacıyla kullanılır.

  !Label Encoding, nominal (sırasız) kategorik değişkenler için uygundur. Eğer kategorik değişkenler arasında bir sıralama (ordinal) ilişkisi varsa, One Hot Encoding yöntemi daha uygun olabilir.



* One Hot Encoding:

  Kategorik değişkenleri ikili (binary) olarak temsil eden bir veri dönüşüm yöntemidir. Bu yöntem, makine öğrenimi modellerinde kategorik değişkenleri
  sayısal değerlere dönüştürmek için kullanılır.Her kategorik değeri yeni ikili (binary) sütunlarla temsil eder.Her kategori için ayrı bir sütun oluşturur ve bu sütunlarda 0 ve 1 değerlerini kullanır.

 -Kategorik özellik sıralı değilse ve kategori sayısı daha azsa etkili bir şekilde uygulanır.
